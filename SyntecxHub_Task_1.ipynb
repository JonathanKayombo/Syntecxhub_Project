```python
import pandas as pd
df = pd.read_csv(r"/content/drive/MyDrive/housing.csv")

df.head()
```

```python
df.dropna(inplace= True)
df.info()
```

```python
from sklearn.model_selection import train_test_split
x = df.drop(['median_house_value'], axis =1)
y = df['median_house_value']

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
train_data = X_train.join(y_train)
train_data
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

# Create a ColumnTransformer to apply OneHotEncoder to 'ocean_proximity'
# and pass through the remaining numerical columns.
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), ['ocean_proximity'])
    ],
    remainder='passthrough'
)

# Apply the preprocessor to X_train
X_train_processed = preprocessor.fit_transform(X_train)

print("X_train_processed shape:", X_train_processed.shape)
```

Now that the categorical features in X_train have been processed using OneHotEncoder, the next logical step is to import and instantiate the LinearRegression model and then fit it to the processed training data (X_train_processed) and the target variable (y_train).



```python
from sklearn.linear_model import LinearRegression

# Instantiate a Linear Regression model
lin_reg = LinearRegression()

# Fit the model to the processed training data and target variable
lin_reg.fit(X_train_processed, y_train)

print("Linear Regression model trained successfully.")
```

To evaluate the model's performance, I need to preprocess the test data (`X_test`) using the same preprocessor fitted on the training data, then make predictions using the trained linear regression model, and finally calculate the RMSE and R2 scores.



```python
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score

# Apply the preprocessor to X_test
X_test_processed = preprocessor.transform(X_test)

# Make predictions on the processed test data
y_pred = lin_reg.predict(X_test_processed)

# Calculate RMSE
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")

# Calculate R-squared (R2) score
r2 = r2_score(y_test, y_pred)
print(f"R-squared (R2) Score: {r2:.2f}")
```

To interpret the model's coefficients, I first need to extract the feature names in the correct order, considering the one-hot encoding applied to 'ocean_proximity' and the remaining numerical features from the `X_train` data, then create a DataFrame to store the coefficients and their corresponding feature names, sort them by absolute value, and finally print the result.



```python
import pandas as pd

# Get feature names for one-hot encoded categorical features
ocean_proximity_features = preprocessor.named_transformers_['cat'].get_feature_names_out(['ocean_proximity'])

# Get names of the remaining numerical features (these are the ones passed through by ColumnTransformer)
numerical_features = [col for col in X_train.columns if col != 'ocean_proximity']

# Combine all feature names in the correct order as they appear in X_train_processed
all_feature_names = list(ocean_proximity_features) + list(numerical_features)

# Create a DataFrame for coefficients
coefficients_df = pd.DataFrame({
    'Feature': all_feature_names,
    'Coefficient': lin_reg.coef_
})

# Sort coefficients by their absolute values to show importance
coefficients_df['Absolute_Coefficient'] = coefficients_df['Coefficient'].abs()
coefficients_df = coefficients_df.sort_values(by='Absolute_Coefficient', ascending=False)

print("Model Coefficients (sorted by absolute value):")
print(coefficients_df.drop(columns='Absolute_Coefficient'))
```
